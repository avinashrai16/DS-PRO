```python
num_col = [col for col in df.columns if df[col].dtype!='object']
```

```python
for col in cat_col :
    plt.figure(figsize=(8,5))
    sns.countplot(x = col, data=df, palette='viridis')
    plt.title(f" Univariate Analysis - {col}")
    plt.show()
```

```python
for col in num_col :
    plt.figure(figsize=(8,5))
    sns.histplot(df[col], kde=True)
    plt.title(f"Univariate Analysis - {col}")
    plt.show()
```
