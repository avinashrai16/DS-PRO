    data interpolation : 
Estimating missing values based on existing data points using interpolation techniques like linear or polynomial interpolation.
1. Linear interpolation
2. cubic interpolation
3. Polynomial interpolation

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1,2,3,4,5])
y = np.array([1,3,5,7,9])
plt.scatter(x,y)
```

```python
# linear interpolation fill the missing data

x_new = np.linspace(1,5,10)
x_new
```

```python
y_interpolation = np.interp(x_new,x,y) # here we r getting the new y based on x_new and original x and y to maintaing the trend
```

```python
plt.scatter(x_new, y_interpolation)
```

```python
# cubic interpolation
x = np.array([1,2,3,4,5])
y = np.array([1,8,27,64,125]) # cubic value for corresponding x value

plt.scatter(x,y) # plot will be kind of curve
```

```python
# cubic interpolation to fill the missing data

from scipy.interpolate import interp1d
```

```python
f = interp1d(x,y,kind='cubic') # function is created to provide cubic relation with x and y 
f
```

```python
x_new = np.linspace(1,5,10)
x_new
```

```python
y_interpolation = f(x_new)
```

```python
y_interpolation
```

```python
plt.scatter(x_new,y_interpolation)
```

```python
# polynomial interpolation
x = np.array([1,2,3,4,5])
y = np.array([1,4,9,1,25])

plt.scatter(x,y) # no clear trend neither linear nor cubic
```

```python
p = np.polyfit(x,y,2) # degree is provided as 2 
```

```python
x_new = np.linspace(1,5,10)
y_interpolation = np.polyval(p,x_new)
```

```python
plt.scatter(x_new,y_interpolation)
```

```python
p = np.polyfit(x,y,3) # degree is provided as 2 
x_new = np.linspace(1,5,10)
y_interpolation = np.polyval(p,x_new)
plt.scatter(x_new,y_interpolation)
```
