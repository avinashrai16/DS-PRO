```python
from sklearn.datasets import make_regression # used for creating regression dataset
from sklearn.model_selection import train_test_split

X, y = make_regression(n_samples=1000, n_features=10, n_targets=1, random_state=1, noise=0.1)
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.20, random_state=1)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
```

```python
from sklearn.ensemble import VotingRegressor
ensemble_reg=VotingRegressor(estimators=[('mlr_reg',LinearRegression()),('support_vec_reg',SVR(kernel="linear")),('decision_tree_reg',DecisionTreeRegressor())])
```

```python
ensemble_reg
```

```python
ensemble_reg.fit(X_train,y_train)
```

```python
y_pred = ensemble_reg.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import r2_score
print(r2_score(y_test,y_pred))
```
