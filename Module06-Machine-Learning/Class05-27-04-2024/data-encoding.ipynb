```python

from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder()
encoder
```

```python
encoder.fit(df[['status']])
```

```python
encoded = encoder.fit_transform(df[['status']]).toarray()
```

```python
encoded
```

```python
encoder.get_feature_names_out()
```

```python
encoder_df = pd.DataFrame(encoded,columns=encoder.get_feature_names_out()) # data set with dummy values
```

```python
 # new data coming in data set after doing the encoding we will use transform
# EX for status as 'Single'

encoder.transform([['Single']]).toarray()
```

```python
# EX for status as 'Married'
encoder.transform([['Married']]).toarray()
```

```python
final_df = pd.concat([df,encoder_df], axis=1)
```

```python
df['qualification']
```

```python
# Target guided ordinal encoding on top of label/ordinal encoding
# relationship along with target variable
# when to use when lot of unique categories
# replace with mean and median

df = pd.DataFrame({'time':['lunch','breakfast','dinner','lunch','breakfast','dinner','lunch','breakfast','dinner'],
              'total_bill':[120,120,40,150,160,120,120,40,150]})
```

```python
df
```

```python
mean_price = df.groupby('time')['total_bill'].mean()
```

```python
mean_price
```

```python
df['time_target_encoded'] = df['time'].map(mean_price)
```

```python
df
```
