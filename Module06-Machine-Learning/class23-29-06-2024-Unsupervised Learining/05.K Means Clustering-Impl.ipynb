```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore")
```

```python
from sklearn.datasets import make_blobs
X, y = make_blobs(n_samples=1000, centers=3, n_features=2, random_state=1)
```

```python
X 
```

```python
y
```

```python
X[:,0]
```

### Y is not required in Unsupervised learning

```python
plt.scatter(X[:,0],X[:,1], c=y) # color is based on y hence 3 color
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=1)
```

```python
X_train.shape, X_test.shape
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3)
```

```python
kmeans
```

```python
kmeans.fit(X_test) # only X_train as this is unsupervised learning
```

```python
y_labels = kmeans.predict(X_test)
```

```python
plt.scatter(X_test[:,0],X_test[:,1], c=y_labels) # color is based on y hence 3 color
```

```python
from sklearn.metrics import silhouette_score
silhouette_score(X_test,y_labels)
```

```python
silhouette_score(X_test,kmeans.labels_)
```

```python
kmeans.inertia_ # for k =3 you have wcss distance
```

## finding the optimum value of k

```python
wcss=[]
for k in range(1,10):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(X_train)
    wcss.append(kmeans.inertia_)
```

```python
wcss
```

# plot an elbow curve

```python
plt.plot(range(1,10),wcss)
# plt.xticks(range(1,10))
plt.xlabel("no of clusters")
plt.ylabel("wcss")
plt.show()
```

## as per the above plot the K optimum value is 3 which is kind of same even

```python
silhoutte_coeff = []
for k in range(2,10):
    kmeans = KMeans(n_clusters = k)
    kmeans.fit(X_train)
    score = silhouette_score(X_train,kmeans.labels_)
    silhoutte_coeff.append(score)
```
