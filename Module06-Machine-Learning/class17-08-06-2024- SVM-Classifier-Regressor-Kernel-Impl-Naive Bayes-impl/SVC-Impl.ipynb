```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# filter warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import make_classification
X, y = make_classification(n_samples=1000, n_features=2, n_classes=2, n_clusters_per_class=2, n_redundant=0)
# need to see the parameters in details 
```

### overlapping of data can be seen below.
### this will be helpful to understand abt SVM , how it handles such data while doing classification

```python
sns.scatterplot(x=pd.DataFrame(X)[0], y=pd.DataFrame(X)[1], hue=y) 
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=1)
```

```python
X_train.shape, X_test.shape
```

```python
y_train.shape, y_test.shape
```

```python
from sklearn.svm import SVC # support vector classifier
classifier = SVC(kernel="linear")
```

```python
classifier
```

```python
classifier.fit(X_train,y_train)
```

```python
classifier.coef_
```

```python
y_pred = classifier.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

print(classification_report(y_test,y_pred))
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
accuracy_score(y_test,y_pred)
```

    HyperParameter tuning

```python
from sklearn.model_selection import GridSearchCV

params={'kernel':['linear'],
          'C':[1,2,3,4,5,10,120,50,100],
          'gamma':[0.2,1,0.4,0.001,0.003]
         }
```

```python
grid = GridSearchCV(SVC(),param_grid=params, cv=5, verbose=3)
grid
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```
