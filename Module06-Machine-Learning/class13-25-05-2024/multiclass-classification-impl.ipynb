```python
from sklearn.datasets import load_iris # get the iris data set
data = load_iris()
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target # in binary classification problem we have filtered one to keep only 2 classes
df.target.unique()
```

```python
X = df.iloc[:, :-1]
y= df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.20, random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression # use logistic regression model as this is a classification problem
classifier = LogisticRegression(max_iter=300) 
# max_iter = This will ensure that the logistic regression model has enough iterations to converge, especially useful for complex datasets.
classifier
```

```python
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
y_pred 
```

```python
classifier.predict_proba(X_test) # below is showing the probability of all 3 classes
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
confusion_matrix(y_test,y_pred)
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```
