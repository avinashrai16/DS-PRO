```python
# Problem 6: Add two numbers represented by linked lists (where each node contains a single digit).
# Input: List 1: 2 -> 4 -> 3, List 2: 5 -> 6 -> 4 (represents 342 + 465)
# Output: 7 -> 0 -> 8 (represents 807)


class ListNode:
    def __init__(self, value=0, next=None):
        self.value = value
        self.next = next

def add_linked_lists(l1, l2):
    dummy_head = ListNode()
    current = dummy_head
    carry = 0

    while l1 or l2 or carry:
        # Extract values of the current nodes or 0 if the node is None
        value1 = l1.value if l1 else 0
        value2 = l2.value if l2 else 0

        # Calculate the sum along with the carry
        total = value1 + value2 + carry
        carry, digit = divmod(total, 10)

        # Create a new node with the calculated digit
        current.next = ListNode(digit)
        current = current.next

        # Move to the next nodes if available
        if l1:
            l1 = l1.next
        if l2:
            l2 = l2.next

    return dummy_head.next

# Example usage:
# Input: List 1: 2 -> 4 -> 3, List 2: 5 -> 6 -> 4 (represents 342 + 465)
# Output: 7 -> 0 -> 8 (represents 807)

# Create linked lists
list1 = ListNode(2, ListNode(4, ListNode(3)))
list2 = ListNode(5, ListNode(6, ListNode(4)))

# Add linked lists
result = add_linked_lists(list1, list2)

# Print the result
while result:
    print(result.value, end=" -> ")
    result = result.next
```

```python
# Problem 7: Swap nodes in pairs in a linked list.
# Input: 1 -> 2 -> 3 -> 4
# Output: 2 -> 1 -> 4 -> 3
class ListNode:
    def __init__(self, value=0, next=None):
      self.value = value
      self.next = next

def swap_pairs(head):
    dummy = ListNode(0)
    dummy.next = head
    current = dummy

    while current.next and current.next.next:
        # Nodes to be swapped
        first_node = current.next
        second_node = current.next.next

        # Perform the swap
        first_node.next = second_node.next
        second_node.next = first_node
        current.next = second_node

        # Move to the next pair
        current = current.next.next

    return dummy.next

# Example usage:
# Input: 1 -> 2 -> 3 -> 4
# Output: 2 -> 1 -> 4 -> 3

# Create a linked list
input_list = ListNode(1, ListNode(2, ListNode(3, ListNode(4))))

# Swap pairs
output_list = swap_pairs(input_list)

# Print the result
while output_list:
    print(output_list.value, end=" -> ")
    output_list = output_list.next

```

```python
# Problem 8: Reverse nodes in a linked list in groups of k.
# Input: 1 -> 2 -> 3 -> 4 -> 5, k = 3
# Output: 3 -> 2 -> 1 -> 4 -> 5

class ListNode:
    def __init__(self, value=0, next=None):
        self.value = value
        self.next = next

def reverse_k_group(head, k):
    def reverse_group(start, end):
        prev, current = None, start
        while current != end:
            temp = current.next
            current.next = prev
            prev = current
            current = temp
        return prev

    dummy = ListNode(0)
    dummy.next = head
    prev_group_end = dummy
    while True:
        # Find the end of the current group
        group_end = prev_group_end
        for _ in range(k):
            if group_end.next:
                group_end = group_end.next
            else:
                return dummy.next

        # Save the next group's start
        next_group_start = group_end.next

        # Reverse the current group
        prev_group_end.next = reverse_group(prev_group_end.next, group_end)

        # Connect the reversed group with the next group
        prev_group_end.next.next = next_group_start

        # Move to the next group
        prev_group_end = prev_group_end.next

# Example usage:
# Input: 1 -> 2 -> 3 -> 4 -> 5, k = 3
# Output: 3 -> 2 -> 1 -> 4 -> 5

# Create a linked list
input_list = ListNode(1, ListNode(2, ListNode(3, ListNode(4, ListNode(5)))))

# Reverse nodes in groups of k
k = 3
output_list = reverse_k_group(input_list, k)

# Print the result
while output_list:
    print(output_list.value, end=" -> ")
    output_list = output_list.next
```

```python
# Problem 9: Determine if a linked list is a palindrome.
# Input: 1 -> 2 -> 2 -> 1
# Output: True
class ListNode:
    def __init__(self, value=0, next=None):
        self.value = value
        self.next = next

def is_palindrome(head):
    def reverse_list(node):
        prev = None
        while node:
            temp = node.next
            node.next = prev
            prev = node
            node = temp
        return prev

    # Find the middle of the linked list
    slow = fast = head
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next

    # Reverse the second half of the linked list
    reversed_second_half = reverse_list(slow)

    # Compare the first half with the reversed second half
    while reversed_second_half:
        if head.value != reversed_second_half.value:
            return False
        head = head.next
        reversed_second_half = reversed_second_half.next

    return True

# Example usage:
# Input: 1 -> 2 -> 2 -> 1
# Output: True

# Create a linked list
input_list = ListNode(1, ListNode(2, ListNode(2, ListNode(1))))

# Check if the linked list is a palindrome
result = is_palindrome(input_list)
print(result)
```

```python
# Problem 10: Rotate a linked list to the right by k places.
# Input: 1 -> 2 -> 3 -> 4 -> 5, k = 2
# Output: 4 -> 5 -> 1 -> 2 -> 3
class ListNode:
    def __init__(self, value=0, next=None):
        self.value = value
        self.next = next

def rotate_right(head, k):
    if not head or k == 0:
        return head

    # Find the length of the linked list
    length = 0
    current = head
    while current:
        length += 1
        current = current.next

    # Calculate the actual rotation value
    actual_k = k % length

    # If actual_k is 0, no rotation is needed
    if actual_k == 0:
        return head

    # Move to the node at position (length - actual_k)
    prev, current = None, head
    for _ in range(length - actual_k):
        prev = current
        current = current.next

    # Break the list at the rotation point
    prev.next = None

    # Connect the rotated part to the beginning
    new_head = current
    while current.next:
        current = current.next
    current.next = head

    return new_head

# Example usage:
# Input: 1 -> 2 -> 3 -> 4 -> 5, k = 2
# Output: 4 -> 5 -> 1 -> 2 -> 3

# Create a linked list
input_list = ListNode(1, ListNode(2, ListNode(3, ListNode(4, ListNode(5)))))

# Rotate the linked list to the right by k places
k = 2
output_list = rotate_right(input_list, k)

# Print the result
while output_list:
    print(output_list.value, end=" -> ")
    output_list = output_list.next


```

```python
# Problem 12: Rearrange a linked list such that all even positioned nodes are placed at the end.
# Input: 1 -> 2 -> 3 -> 4 -> 5
# Output: 1 -> 3 -> 5 -> 2 -> 4
class ListNode:
    def __init__(self, value=0, next=None):
        self.value = value
        self.next = next

def rearrange_linked_list(head):
    if not head or not head.next or not head.next.next:
        return head

    # Separate odd and even positioned nodes
    odd_head = ListNode(0)
    even_head = ListNode(0)
    odd_current = odd_head
    even_current = even_head
    is_odd = True

    current = head
    while current:
        if is_odd:
            odd_current.next = current
            odd_current = odd_current.next
        else:
            even_current.next = current
            even_current = even_current.next

        is_odd = not is_odd
        current = current.next

    # Combine the odd and even positioned lists
    odd_current.next = even_head.next
    even_current.next = None  # Set the end of even list to None

    return odd_head.next

# Example usage:
# Input: 1 -> 2 -> 3 -> 4 -> 5
# Output: 1 -> 3 -> 5 -> 2 -> 4

# Create a linked list
input_list = ListNode(1, ListNode(2, ListNode(3, ListNode(4, ListNode(5)))))

# Rearrange the linked list
output_list = rearrange_linked_list(input_list)

# Print the result
while output_list:
    print(output_list.value, end=" -> ")
    output_list = output_list.next
```

```python
# Problem 13: Given a non-negative number represented as a linked list, add one to it.
# Input: 1 -> 2 -> 3 (represents the number 123)
# Output: 1 -> 2 -> 4 (represents the number 124)

class ListNode:
    def __init__(self, value=0, next=None):
        self.value = value
        self.next = next

def add_one_to_linked_list(head):
    # Step 1: Reverse the linked list
    def reverse_list(node):
        prev = None
        while node:
            temp = node.next
            node.next = prev
            prev = node
            node = temp
        return prev

    # Step 2: Add one to the reversed linked list
    reversed_head = reverse_list(head)
    current = reversed_head
    carry = 1  # Start with a carry of 1

    while current and carry:
        total = current.value + carry
        carry, current.value = divmod(total, 10)
        current = current.next

    # If there is still a carry, create a new node
    if carry:
        reversed_head = ListNode(carry, reversed_head)

    # Step 3: Reverse the linked list again
    result_head = reverse_list(reversed_head)

    return result_head

# Example usage:
# Input: 1 -> 2 -> 3 (represents the number 123)
# Output: 1 -> 2 -> 4 (represents the number 124)

# Create a linked list
input_list = ListNode(1, ListNode(2, ListNode(3)))

# Add one to the linked list
output_list = add_one_to_linked_list(input_list)

# Print the result
while output_list:
    print(output_list.value, end=" -> ")
    output_list = output_list.next


```

```python
# Problem 15: Find the minimum element in a rotated sorted array.
# Input: [4, 5, 6, 7, 0, 1, 2]
# Output: 0

def find_min(nums):
    left, right = 0, len(nums) - 1

    while left < right:
        mid = left + (right - left) // 2

        if nums[mid] > nums[right]:
            left = mid + 1
        else:
            right = mid

    return nums[left]


nums = [4, 5, 6, 7, 0, 1, 2]
result = find_min(nums)
print(result)

```

```python
# Problem 16: Search for a target value in a rotated sorted array.
# Input: nums = [4, 5, 6, 7, 0, 1, 2], target = 0
# Output: 4

def search_rotated_array(nums, target):
    left, right = 0, len(nums) - 1

    while left <= right:
        mid = left + (right - left) // 2

        if nums[mid] == target:
            return mid

        # Check which part is sorted
        if nums[left] <= nums[mid]:
            # Left part is sorted
            if nums[left] <= target < nums[mid]:
                right = mid - 1
            else:
                left = mid + 1
        else:
            # Right part is sorted
            if nums[mid] < target <= nums[right]:
                left = mid + 1
            else:
                right = mid - 1

    return -1  # Target not found


nums = [4, 5, 6, 7, 0, 1, 2]
target = 0
result = search_rotated_array(nums, target)
print(result)
```

```python
# Problem 17: Find the peak element in an array. A peak element is greater than its neighbors.
# Input: nums = [1, 2, 3, 1]
# Output: 2 (index of peak element)
def find_peak_element(nums):
    left, right = 0, len(nums) - 1

    while left < right:
        mid = left + (right - left) // 2
        # print(nums[mid])
        # Check if mid is a peak
        if nums[mid] > nums[mid + 1]:
            right = mid
        else:
            left = mid + 1

    return left  # left is the index of the peak element


nums = [1, 2, 3, 1]
result = find_peak_element(nums)
print(result)
```

```python
# Problem 18: Given a m x n matrix where each row and column is sorted in ascending order, count the number
# of negative numbers.
# Input: grid = [[4, 3, 2, -1], [3, 2, 1, -1], [1, 1, -1, -2], [-1, -1, -2, -3]]
# Output: 8

def count_negatives(grid):
    m, n = len(grid), len(grid[0])
    count = 0
    row, col = 0, n - 1

    while row < m and col >= 0:
        if grid[row][col] < 0:
            count += m - row
            col -= 1
        else:
            row += 1

    return count

grid = [
    [4, 3, 2, -1],
    [3, 2, 1, -1],
    [1, 1, -1, -2],
    [-1, -1, -2, -3]
]
result = count_negatives(grid)
print(result)
```

```python
# Problem 19: Given a 2D matrix sorted in ascending order in each row, and the first integer of each row is
# greater than the last integer of the previous row, determine if a target value is present in the matrix.
# Input: matrix = [[1, 3, 5, 7], [10, 11, 16, 20], [23, 30, 34, 60]], target = 3
# Output: True

def search_matrix(matrix, target):
    if not matrix or not matrix[0]:
        return False

    m, n = len(matrix), len(matrix[0])
    left, right = 0, m * n - 1

    while left <= right:
        mid = left + (right - left) // 2
        mid_value = matrix[mid // n][mid % n]

        if mid_value == target:
            return True
        elif mid_value < target:
            left = mid + 1
        else:
            right = mid - 1

    return False

matrix = [
    [1, 3, 5, 7],
    [10, 11, 16, 20],
    [23, 30, 34, 60]
]
target = 3
result = search_matrix(matrix, target)
print(result)
```

```python
# Problem 20: Find Median in Two Sorted Arrays
# Problem: Given two sorted arrays, find the median of the combined sorted array.
# Input: nums1 = [1, 3], nums2 = [2]
# Output: 2.0

def findMedianSortedArrays(nums1, nums2):
    if len(nums1) > len(nums2):
        nums1, nums2 = nums2, nums1

    x, y = len(nums1), len(nums2)
    low, high = 0, x

    while low <= high:
        partitionX = (low + high) // 2
        partitionY = (x + y + 1) // 2 - partitionX

        maxX = float('-inf') if partitionX == 0 else nums1[partitionX - 1]
        minX = float('inf') if partitionX == x else nums1[partitionX]

        maxY = float('-inf') if partitionY == 0 else nums2[partitionY - 1]
        minY = float('inf') if partitionY == y else nums2[partitionY]

        if maxX <= minY and maxY <= minX:
            if (x + y) % 2 == 0:
                return (max(maxX, maxY) + min(minX, minY)) / 2.0
            else:
                return max(maxX, maxY)
        elif maxX > minY:
            high = partitionX - 1
        else:
            low = partitionX + 1


nums1 = [1, 3]
nums2 = [2]
result = findMedianSortedArrays(nums1, nums2)
print(result)
```

```python
# Problem 21: Given a sorted character array and a target letter, find the smallest letter in the array that is
# greater than the target.
# Input: letters = ['c', 'f', 'j'], target = a
# Output: 'c'

def next_greatest_letter(letters, target):
    left, right = 0, len(letters)

    while left < right:
        mid = left + (right - left) // 2

        if letters[mid] <= target:
            left = mid + 1
        else:
            right = mid

    # If left is equal to the length of the array, wrap around to the beginning
    return letters[left % len(letters)]

# Example usage:
letters = ['c', 'f', 'j']
target = 'a'
result = next_greatest_letter(letters, target)
print(result)
```

```python
# Problem 22: Given an array with n objects colored red, white, or blue, sort them in-place so that objects of
# the same color are adjacent, with the colors in the order red, white, and blue.
# Input: nums = [2, 0, 2, 1, 1, 0]
# Output: [0, 0, 1, 1, 2, 2]
def sortColors(nums):
    low, mid, high = 0, 0, len(nums) - 1

    while mid <= high:
        if nums[mid] == 0:
            nums[low], nums[mid] = nums[mid], nums[low]
            low += 1
            mid += 1
        elif nums[mid] == 1:
            mid += 1
        else:
            nums[mid], nums[high] = nums[high], nums[mid]
            high -= 1

# usage:
nums = [2, 0, 2, 1, 1, 0]
sortColors(nums)
print(nums)
```

```python
# Problem 23: Find the kth largest element in an unsorted array.
# Input: nums = [3, 2, 1, 5, 6, 4], k = 2
# Output: 5

def findKthLargest(nums, k):
    nums.sort(reverse=True)
    return nums[k - 1]


nums = [3, 2, 1, 5, 6, 4]
k = 2
result = findKthLargest(nums, k)
print(result)
```

```python
# Problem 24: Given an unsorted array, reorder it in-place such that nums[0] <= nums[1] >= nums[2] <=nums[3]...
# Input: nums = [3, 5, 2, 1, 6, 4]
# Output: [3, 5, 1, 6, 2, 4]

def wiggleSort(nums):
    for i in range(len(nums) - 1):
        if i % 2 == 0 and nums[i] > nums[i + 1]:
            nums[i], nums[i + 1] = nums[i + 1], nums[i]
        elif i % 2 == 1 and nums[i] < nums[i + 1]:
            nums[i], nums[i + 1] = nums[i + 1], nums[i]


nums = [3, 5, 2, 1, 6, 4]
wiggleSort(nums)
print(nums)
```

```python
# Problem 26: Find the maximum element in an array of integers.
# Input: [3, 7, 2, 9, 4, 1]
# Output: 9

def get_max(list1):
  max_element = None
  for i in list1:
    if max_element:
      if max_element<i:
        max_element = i
    else:
      max_element = i
  return max_element

print(get_max([3, 7, 2, 9, 4, 1]))

```

```python
# Problem 27: Implement linear search to find the index of a target element in an array.
# Input: [5, 3, 8, 2, 7, 4], target = 8
# Output: 2

def linear_search(arr, target):
    for i in range(len(arr)):
        if arr[i] == target:
            return i
    return -1  # Target not found

arr = [5, 3, 8, 2, 7, 4]
target = 8
result = linear_search(arr, target)

if result != -1:
    print(f"Element {target} found at index {result}.")
else:
    print(f"Element {target} not found in the array.")
```

```python
# Problem 28 Calculate the factorial of a given number.
# Input: 5
# Output: 120 (as 5! = 5 * 4 * 3 * 2 * 1 = 120)


def get_factorial(num):
  if num == 1:
    return 1
  return num*get_factorial(num-1)

print(get_factorial(5))
```

```python
# Problem 29: Check if a given number is a prime number.
# Input: 7
# Output: True
def is_prime(num):
    if num < 2:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True


number = 7
if is_prime(number):
    print(f"{number} is a prime number.")
else:
    print(f"{number} is not a prime number.")

```

```python
# Problem 30: Generate the Fibonacci series up to a given number n.
# Input: 8
# Output: [0, 1, 1, 2, 3, 5, 8, 13]

def generate_fibonacci(n):
    fibonacci_series = []
    a, b = 0, 1
    while len(fibonacci_series) < n:
        fibonacci_series.append(a)
        a, b = b, a + b
    return fibonacci_series

n = 8
fibonacci_result = generate_fibonacci(n)
print(f"Fibonacci series up to {n}: {fibonacci_result}")
```
