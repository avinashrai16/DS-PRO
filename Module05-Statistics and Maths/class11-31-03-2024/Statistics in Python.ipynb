```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# normal/gaussion distribution
# mean and std deviation 
mu, sigma = 0, 0.1

s = np.random.normal(mu, sigma,size=10000)
print("count of data points",len(s))
s
```

```python
plt.hist(s,bins=40, density=True)
plt.show()
```

```python
sns.distplot(s)
```

```python
# uniform distribution > all the outcome has equal probability to occur in range

uni = np.random.uniform(0,1,1000)
plt.hist(uni)
```

```python
# log normal distribution >> where logarithm is normally distributed

log = np.random.lognormal(mean=0, sigma=0.5, size=1000)
log = log[np.where(abs(log)<5)]

plt.hist(log, bins=100)
plt.show()
```

```python
# convert log normal to normal distribution >> transformation
sns.displot(np.log(log),kde=True,kind='hist')
```

```python
sns.displot(np.sqrt(log),kde=True,kind='hist')
```

```python
sns.displot(np.square(log),kde=True,kind='hist')
```

```python
# chi square distribution

chi = np.random.chisquare(df=1, size=10000)
chi = chi[np.where(abs(chi)<10)]
plt.hist(chi, bins=100)
plt.show()
```

```python
chi = np.random.chisquare(df=5, size=10000)
chi = chi[np.where(abs(chi)<10)]
plt.hist(chi, bins=100)
plt.show()
```

```python
# T distribution

t = np.random.standard_t(df=1, size=10000)
t = t[np.where(abs(t)<10)]
plt.hist(t, bins=100)
plt.show()
```

```python
# f distribution
# dfnum is for numerator
# dfden is for denominator

f = np.random.f(dfnum=1, dfden=0.5, size=10000)
f = f[np.where(abs(f)<10)]
plt.hist(f, bins=100)
plt.show()
```

```python
# binomial distribution with 1 trail , n bernoulli trail is binomial
b = np.random.binomial(n=1, p=0.5, size=1000)
plt.hist(b)
plt.show()
```

```python
b = np.random.binomial(n=100, p=0.5, size=1000)
plt.hist(b)
plt.show()
```

```python
p = np.random.poisson(lam=5, size=1000)
plt.hist(p, bins=20)
plt.show()
```
