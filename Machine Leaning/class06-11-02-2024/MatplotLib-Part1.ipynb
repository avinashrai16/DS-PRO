```python
x = np.linspace(0, 10, 100)
```

```python
x
```

```python
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
```

```python
plt.plot(x, np.sin(x),'--')
```

```python
plt.plot(x, np.sin(x),'.')
```

```python
plt.figure() # make a canvas
plt.subplot(1,2,1) # on the step 1 page , make sub plot and 2 plots and access the 1st plot
plt.plot(x, np.sin(x))

plt.subplot(1,2,2) # on the step 1 page , make sub plot and 2 plots and access the 2nd plot
plt.plot(x, np.cos(x))

plt.show()
```

```python
plt.figure() # make a canvas
plt.subplot(2,2,1) # nrows, ncols:  create 2*2 plot and access the 1st plot
plt.plot(x, np.sin(x))

plt.subplot(2,2,2) # nrows, ncols:  create 2*2 plot and access the 2nd plot
plt.plot(x, np.cos(x))


plt.subplot(2,2,3) # nrows, ncols:  create 2*2 plot and access the 3rd plot
plt.plot(x, np.tan(x))

plt.subplot(2,2,4) # nrows, ncols:  create 2*2 plot and access the 4th plot
plt.plot(x, np.cos(x+1))

plt.show()
```

```python
plt.figure() # make a canvas
plt.subplot(2,1,1) # # nrows, ncols: and access the 1st plot
plt.plot(x, np.sin(x))

plt.subplot(2,1,2) # # nrows, ncols: and access the 2nd plot
plt.plot(x, np.cos(x))

plt.show()
```

```python
plt.plot(x, np.sin(x-0), color='red')
plt.plot(x, np.sin(x-1), color='green')
plt.plot(x, np.sin(x-2), color=(1.0,0.5,0)) # rgb
```

```python
# can we change the values of x and y axis 
plt.plot(x, np.sin(x-0), color='blue')
plt.ylim(-1.5,1.5)
plt.xlim(1,15)
plt.show()
```

```python
# can we change the plot to cover all the empty space above 
plt.plot(x, np.sin(x-0), color='blue')
plt.ylim(-1.5,1.5)
plt.xlim(1,15)
plt.axis("equal")
plt.show()
```

```python
plt.figure() # make a canvas
plt.subplot(2,1,1) # on the step 1 page , make sub plot and 2 plots and access the 1st plot
plt.plot(x, np.sin(x-0), color='blue')
plt.ylim(-1.5,1.5)
plt.xlim(1,15)
plt.axis("equal")
plt.show()
```

```python
# how to put grid on the graph, put title, x label y label

plt.plot(x, np.sin(x-0), color='blue')
plt.ylim(-1.5,1.5)
plt.xlim(1,15)
plt.grid()
plt.title("sample title")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.axis("equal")
plt.show()
```

```python
# how to put grid on the graph, put title, x label y label

plt.plot(x, np.sin(x-0), color='blue', label = "Sinx")
plt.plot(x, np.cos(x-0), color='green', label = "Cosx")
plt.ylim(-1.5,1.5)
plt.xlim(1,15)
plt.grid()
plt.title("sample title")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.axis("equal")
plt.legend()
plt.show()
```
