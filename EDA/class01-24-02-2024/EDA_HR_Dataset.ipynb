```python
df = pd.read_csv("HRDataset_v14.csv")
```

```python
# compare top 10 highest and lowest paid employee

c =[i for i in range (1,51)]
x = df['Salary'].sort_values(ascending=False).head(50)
y = df['Salary'].sort_values(ascending=False).tail(50)
plt.figure(figsize=(15, 6))  
plt.bar(c,x,color='g',label="highest_salary")
plt.bar(c,y,color='r',label="lowest_salary")
plt.title("Top 10 highest and lowest comparison", fontsize=15)
plt.xticks(c)
# plt.xticks(rotation=90)
plt.ylabel('Salaries')
plt.legend()
plt.show()
```
